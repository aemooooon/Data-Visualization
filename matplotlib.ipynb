## 折线图 plot

> 某指标随时间的变化呈现的变化规律

### 基本思路
1. 准备X（时间），Y（温度）
2. 创建画板
3. 绘制图像
4. 显示图像

### Marker
* 's' : 方块状
* 'o' : 实心圆
* '^' : 正三角形
* 'v' : 反正三角形
* '+' : 加好
* '*' : 星号
* 'x' : x号
* 'p' : 五角星
* '1' : 三脚架标记
* '2' : 三脚架标记

````python
import matplotlib.pyplot as plt
import random
%matplotlib inline

# 创建画布
# figsize 图像宽高 dpi 图像清晰度
plt.figure(figsize=(6, 3), dpi=200)

# 把数据画出来
plt.plot([1, 2, 3, 4, 5, 6, 7], [12, 14, 15, 17, 14, 10, 13], marker='*', color='r', ms=10) # 某城市一周的温度变化

# 保存图像到当前文件夹
plt.savefig("./tmp/test.png")

# 显示图像
plt.show()

# 常用参数
# figsize=(15,7.5), dpi= 80
# figsize=(12,6)  , dpi=100
# figsize=( 8,4)  , dpi=150
# figsize=( 6,3)  , dpi=200
````

### 画出某城市11点到12点1小时内每分钟的温度变化折线图，温度在15-18度之间__

````python
x = range(60)
y_guangzhou = [random.uniform(15,20) for i in x]
y_haerbing = [random.uniform(0,5) for i in x]

plt.figure(figsize=(12,6), dpi=100)
plt.plot(x, y_guangzhou, color='r', linestyle=':', label="广州")
plt.plot(x, y_haerbing, linestyle='--', label="哈尔滨")
x_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5], x_label[::5])
plt.yticks(range(0, 30, 5))
plt.grid(True, linestyle='--', alpha=0.3) # 网格属性
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("18日上午11点到12点哈尔滨和广州的气温变化图")
plt.legend(loc="upper left") # define inside of the plot function as label

plt.show()
````

### 一个平面多个插线图

````python
x = range(60)
y_guangzhou = [random.uniform(15,20) for i in x]
y_haerbing = [random.uniform(0,5) for i in x]

figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,6), dpi=100)

axes[0].plot(x, y_guangzhou, color='r', linestyle='--', label="广州")
axes[1].plot(x, y_haerbing, color='b', linestyle=':', label="哈尔滨")

x_label = ["11点{}分".format(i) for i in x]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_label)
axes[0].set_yticks(range(0, 40, 5))

axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_label)
axes[1].set_yticks(range(0, 40, 5))

axes[0].grid(True, linestyle='--', alpha=0.3) # 网格属性
axes[1].grid(True, linestyle='--', alpha=0.3)

axes[0].set_xlabel("时间")
axes[0].set_ylabel("温度")
axes[0].set_title("18日上午11点到12点广州的气温变化图")

axes[1].set_xlabel("时间")
axes[1].set_ylabel("温度")
axes[1].set_title("18日上午11点到12点哈尔滨的气温变化图")

axes[0].legend(loc="upper left") # define inside of the plot function as label
axes[1].legend(loc="upper left")

plt.show()
````

## 柱状图 bar

> 分别显示或对比各个对象的数据大小

### Example 1

````python
np.random.seed(0)
x = ['A','B', 'C', 'D', 'E', 'F', 'G']
y = np.random.randint(1, 100000, 7)

x_ticks = range(len(x))

plt.figure(figsize=(8,4), dpi=150)
plt.bar(x, y, color=['r', 'g', 'b', 'c', 'm', 'y', 'k'])
plt.xticks(x_ticks, x)
plt.grid(linestyle='--', alpha=0.3)
plt.title("票房对比图")
plt.show()
````

### Example 2

````python
np.random.seed(0)
x = ['A','B', 'C', 'D', 'E', 'F', 'G']
y = np.random.randint(1, 100000, 7)
y_firstday = np.random.randint(1, 20000, 7)

x_ticks = range(len(x))

plt.figure(figsize=(12,6), dpi=100)
plt.bar(x_ticks, y, width=0.3, label='First Weekend', color='m')
plt.bar([i+0.3 for i in x_ticks], y_firstday, width=0.3, label="First day", color="r")
plt.xticks([i+0.15 for i in x_ticks], x)
plt.grid(linestyle='--', alpha=0.3)
plt.title("票房对比图")
plt.legend()
plt.show()
````

## 直方图 histogram

> 变量的分布规律, 集中分布，异常值

* 横坐标通常是连续的数字区间，而柱状图通常是类别（之间没有联系）
* 众数，中数，平均数

````python
from scipy import stats
np.random.seed(250)
times = np.random.randint(90, 180, 250)

# bins 公式（用最大值减去最小值除以频距）

plt.figure(figsize=(20, 10), dpi=60)
plt.hist(times, bins=(max(times) - min(times)) // 2, color="c", density=True) # density True(显示频率) False(显示分布数量)
plt.xticks(range(min(times), max(times), 2))
plt.grid(linestyle='--', alpha=0.3)
plt.title('电影时长分布情况')
plt.axvline(np.mean(times), color='r')
plt.axvline(np.median(times), color='g')
plt.axvline(stats.mode(times)[0][0], color='b', linewidth='5')
plt.show()
````

## 饼图 pie

> 几个对象的占比情况

````python
np.random.seed(0)
names = ['东邪','西毒', '南帝', '北丐', '中神通', '神雕侠侣', '铁掌水上漂']
x = np.random.randint(1, 100000, 7)

plt.figure(figsize=(7, 7), dpi=120)
plt.pie(x, labels=names, colors=['r', 'g', 'b', 'c', 'm', 'y', 'c'], autopct='%1.2f%%')
plt.title('电影票房占比图')
plt.axis('equal') # 保持图形比例
plt.legend(loc=8)
plt.show()
````

### 疫情历史数据走势图

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_json('./json/history.json') # http://www.dzyong.top:3005/yiqing/history
df = df["data"]
````

````python
x = []
y_confirmed = []
y_suspected = []
y_curesNum = []
y_deathsNum = []

for i in range(df.size):
    x.append(df[i]["date"])
    y_confirmed.append(df[i]["confirmedNum"])
    y_suspected.append(df[i]["suspectedNum"])
    y_curesNum.append(df[i]["curesNum"])
    y_deathsNum.append(df[i]["deathsNum"])

plt.figure(figsize=(8, 4), dpi=150)
plt.plot(x, y_confirmed, linestyle="-", color='r', label="确诊")
plt.plot(x, y_suspected, linestyle="-", color='y', label="疑似")
plt.plot(x, y_curesNum, linestyle="-", color='g', label="治愈")
plt.plot(x, y_deathsNum, linestyle="-", color='k', label="死亡")
x_label = ["{}".format(i[5:]) for i in x]
plt.xticks(x[::3], x_label[::3])
plt.grid(linestyle="--", alpha=0.3)
plt.title('新型管状病毒发展趋势自2020年1月13日')
plt.xlabel("日期")
plt.ylabel("数量")
plt.legend()
plt.show()
````
